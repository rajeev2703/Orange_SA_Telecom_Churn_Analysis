## ***1. State-wise distribution of revenue***


```python
# Creating a New Coloumn, Total_Revenue, that sums up Total Day Charge, Night Charge, Evening Charge and International charges
def total_rev(a,b,c,d):
  return(a+b+c+d)
df["Total_revenue"]=df.apply(lambda x:total_rev(x["TotalDayCharge"],x["TotalEveCharge"],x["TotalNightCharge"],x["TotalIntlCharge"]),axis=1)
df.head(3)
```

```python
df_new=pd.DataFrame(df.groupby(by=["State"],as_index=False)["Total_revenue"].sum())

plt.rcParams["figure.figsize"] = [17, 7]
sns.set_style('darkgrid')
vis1=sns.barplot(x="State",y="Total_revenue",data=df_new)
plt.xticks(rotation=90,fontsize=14)
plt.yticks(fontsize=15)
plt.title("Revenue Distribution accross Various States",size=15)
plt.ylabel("Revenue ",size=15,labelpad=20)
plt.xlabel("State",size=18,labelpad=20)
plt.legend(prop={"size":18})
plt.show()
```

1. Highest revenue comes in from the states of WV(West Virginia), NY(Newyork)
2. But surprisingly, state of CA(California) has lowest revenue yield, Efforts should be made to boost sales and yield revenue from CA 

## ***2.  Distribution of users who have cancelled the Subscription and those who have stayed***

## ***4.  Determining number of churned and not churned users on the basis of Customer Service Calls***

```python
#Creating a dataframe to calculate the churned users for different number of Customer Service Calls
churn_true=pd.DataFrame(df[df["Churn"]==True].groupby("CustomerServiceCalls",as_index=False)["Churn"].count())
churn_true.rename(columns={"Churn":"ChurnCount"},inplace=True)

#Creating a dataframe to calculate the NonChurned users for different number of Customer Service Calls
churn_false=pd.DataFrame(df[df["Churn"]==False].groupby("CustomerServiceCalls",as_index=False)["Churn"].count())
churn_false.rename(columns={"Churn":"NotChurnCount"},inplace=True)

#Calculating the total number of churned and non churned users
total_churned_users=churn_true["ChurnCount"].sum()
total_not_churned_users=churn_false["NotChurnCount"].sum()

#Calculating the percentage of churned and non-churned users 
churn_true["Churned"]=(churn_true["ChurnCount"]/(total_churned_users))*100
churn_false["NotChurned"]=(churn_false["NotChurnCount"]/(total_not_churned_users))*100

#Creating a new dataframe by merging churn_true and churn_false dataframes
new_df=pd.merge(churn_false,churn_true,on="CustomerServiceCalls")
df_churn=new_df[["CustomerServiceCalls","Churned","NotChurned"]]
#plotting the data
df_churn.set_index('CustomerServiceCalls').plot(kind='bar',stacked=False,color=['magenta','aqua'])
plt.xticks(rotation=0)
plt.ylabel("Churned and Non-Churned percentage ",size=15,labelpad=20)
plt.xlabel("Number of Customer Service calls",size=18,labelpad=20)
plt.legend(prop={"size":15})
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

1. It can be clearly observed that the customer retention is higher with decreased number of Customer Service call. Effectively at 1 call.
2. As the number of calls increased from 3, the percentage of people who cancelled their subscription starts increasing rapidly.
Hence, more effort should be made towards decreasing the no of service calls by providing adequate training to service representatives.

## ***6.  What rates are being charged for the day, evening, and night users? , and how do they compare to rates charged for international calls?***

```python
# Adding new columns "ChargePerMinute" for Various slots, that tells how much the user is being charged per minute for different categories.
df["IntlChargePerMinute"]=df["TotalIntlCharge"]/df['TotalIntlMinutes']
df["DayChargePerMinute"]=df["TotalDayCharge"]/df['TotalDayMinutes']
df["NightChargePerMinute"]=df["TotalNightCharge"]/df['TotalNightMinutes']
df["EveChargePerMinute"]=df["TotalEveCharge"]/df['TotalEveMinutes']
mean_day_charge = df["DayChargePerMinute"].mean()
mean_night_charge = df["NightChargePerMinute"].mean()
mean_eve_charge = df["EveChargePerMinute"].mean()
mean_Intl_charge = df["IntlChargePerMinute"].mean()

plt.rcParams["figure.figsize"] = [15, 7]
labels=["Day","Evening", "Night", "International"]
values=[mean_day_charge, mean_night_charge, mean_eve_charge, mean_Intl_charge]
plt.bar(labels,values, width=0.5)
plt.title("Comparison of Day, Evening, Night, International Call with Charges per minute",size=20)
plt.xlabel("Call Time Slots", size=15)
plt.ylabel("Charge Per Minute", size=15)
plt.show()
```

1. Here its evident that the day charges per minute is relatively higher than evening and night charges.
2. The prices should be kept identical in all slots of time so that customers won't churn away.
